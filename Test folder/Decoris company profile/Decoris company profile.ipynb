```python
def label_satisfactionrating(row):
    
#   extracting Satisfaction rating
    if '"question":"Overall Satisfaction Rate","answer":"Good"' in row ['satisfactionSurvey.data'] :
        return 'Good'
    elif '"question":"Overall Satisfaction Rate","answer":"Great"' in row ['satisfactionSurvey.data'] :
        return 'Great'
    elif '"question":"Overall Satisfaction Rate","answer":"Bad"' in row ['satisfactionSurvey.data'] :
        return 'Bad'
    elif '"question":"Overall Satisfaction Rate","answer":"Terrible"' in row ['satisfactionSurvey.data'] :
        return 'Terrible'
    else:
        return 'No Response'
    
   
```

```python
data['satisfaction_rating'] = data.apply(lambda row: label_satisfactionrating(row), axis=1)
```

```python
import re

s= '[{"department":"dm","user":{"$oid":"6108f12ed9d96600098fbfd3"}},{"department":"commercial","user":{"$oid":"5fe94a4b73981c0009d84195"}},{"department":"creative","user":{"$oid":"5fe9481fb5b8e70008c390da"}}]'
#This is to extract exact ID 
result = re.search('dm(.*?)}', s)
# print(result.group(1))
# result.group(1)
re.search('dm(.*?)}', s).group(1)
```

```python
def extract_id_commercial (col):    
    #   Extracting exact commercial PIC ID
        return re.findall('commercial(.*?)}',col)

def extract_id_dm (col):    
    #   Extracting exact dm ID
        return re.findall('dm(.*?)}',col)

def extract_id_creative (col):
    #   Extracting exact creative ID
        return re.findall('creative(.*?)}', col)
```

```python
data['id_commercial'] = data['PICs'].apply(lambda row: extract_id_commercial(row))
```

```python
data['id_dm']= data['PICs'].apply(lambda row: extract_id_dm(row))
```

```python
data['id_creative']= data['PICs'].apply(lambda row: extract_id_creative(row))
```
